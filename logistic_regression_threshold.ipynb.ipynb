```python
# Step 1: Import required libraries
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Define logistic function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Step 3: Set your inputs
x = np.arange(0, 11, 1)        # X values from 0 to 10
b0 = -6
b1 = 1.2

# Step 4: Calculate z = b0 + b1*x
z = b0 + b1 * x

# Step 5: Apply sigmoid to get ŷ
y_hat = sigmoid(z)

# Step 6: Apply threshold
y_pred = [1 if prob >= 0.5 else 0 for prob in y_hat]

# Step 7: Print results
for i in range(len(x)):
    print(f"x={x[i]:>2}  z={z[i]:>5.2f}  ŷ={y_hat[i]:.4f}  class={y_pred[i]}")
```

```python
# Plot sigmoid curve with threshold
plt.figure(figsize=(8, 5))
plt.plot(x, y_hat, 'bo-', label="Logistic Prediction")
plt.axhline(y=0.5, color='red', linestyle='--', label="Threshold = 0.5")
plt.xlabel("X values")
plt.ylabel("Predicted Probability")
plt.title("Logistic Regression Sigmoid Curve")
plt.grid()
plt.legend()
plt.show()
```
